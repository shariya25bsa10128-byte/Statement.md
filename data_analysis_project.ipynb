```python
import matplotlib.pyplot as plt

department_salary = data.groupby("Department")["Monthly_Salary"].mean()

department_salary.plot(kind="bar")

plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(data['Experience_Years'], data['Monthly_Salary'])

plt.title('Experience vs Monthly Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Monthly Salary')

plt.show()
```

# Employee Salary Data Analysis

## Objective

To analyze an employee salary dataset using Pandas and create visualizations using Matplotlib and Seaborn.

## Tools Used

* Python
* Pandas
* Matplotlib
* Seaborn
* Google Colab

## Analysis Performed

1. Loaded the employee salary dataset from a CSV file using Pandas.
2. Explored the dataset structure and columns.
3. Calculated salary statistics such as average, maximum, and minimum salary.
4. Created visualizations including:

   * Bar Chart
   * Scatter Plot
   * Heatmap

## Results

* Average Monthly Salary: ₹82,288.80
* Maximum Monthly Salary: ₹149,123
* Minimum Monthly Salary: ₹28,420

## Observations

1. The Marketing department has the highest average salary among all departments.
2. The average employee salary in the dataset is ₹82,288.80.
3. The highest salary recorded is ₹149,123, while the lowest salary is ₹28,420.
4. The scatter plot indicates that employees with more years of experience generally tend to earn higher salaries.
5. The heatmap shows a positive relationship between experience and salary, suggesting that salary tends to increase with experience.
6. Salary distribution varies across departments, indicating that department plays an important role in determining employee compensation.

## Conclusion

The employee salary dataset was successfully analyzed using Python and Pandas. Visualizations created with Matplotlib and Seaborn helped identify salary patterns and relationships between variables. The analysis shows that department and experience are important factors affecting employee salaries.
